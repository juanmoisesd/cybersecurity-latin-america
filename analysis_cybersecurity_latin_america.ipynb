Analysis of cyberattacks, ransomware incidents, ITU Global Cybersecurity Index (GCI), and digital infrastructure indicators across Latin America.

````python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import seaborn as sns

plt.rcParams['figure.dpi'] = 120
sns.set_theme(style='darkgrid')
print('Libraries loaded ✓')
````

## 1. Global Cybersecurity Index (GCI-ITU) and Ransomware Attacks

````python
data = {
    'Country': ['Brazil','Mexico','Colombia','Argentina','Chile','Peru','Ecuador','Costa Rica','Uruguay','Venezuela'],
    'GCI_2020': [70.6, 63.4, 57.2, 55.8, 53.7, 49.3, 38.2, 51.1, 48.9, 28.4],
    'GCI_2024': [84.5, 78.3, 72.1, 69.8, 67.4, 58.2, 52.6, 65.3, 61.7, 35.2],
    'Ransomware_2022': [3456, 2891, 1203, 987, 654, 432, 321, 289, 198, 156],
    'Ransomware_2024': [4120, 3340, 1456, 1123, 789, 534, 412, 356, 245, 189],
    'Internet_pct_2024': [84.2, 76.3, 72.4, 88.5, 91.2, 67.8, 65.4, 89.1, 93.7, 34.6]
}
df = pd.DataFrame(data).set_index('Country')
df
````

````python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# GCI comparison
ax1 = axes[0]
x = range(len(df))
w = 0.35
ax1.bar([i-w/2 for i in x], df['GCI_2020'], w, label='GCI 2020', color='#3498db', alpha=0.8)
ax1.bar([i+w/2 for i in x], df['GCI_2024'], w, label='GCI 2024', color='#2ecc71', alpha=0.8)
ax1.set_title('Global Cybersecurity Index (GCI-ITU)\n2020 vs 2024', fontsize=13, fontweight='bold')
ax1.set_xticks(list(x))
ax1.set_xticklabels(df.index, rotation=45, ha='right', fontsize=9)
ax1.set_ylabel('GCI Score (0–100)')
ax1.legend()
ax1.set_ylim(0, 100)
ax1.grid(True, alpha=0.3, axis='y')

# Ransomware vs Internet penetration scatter
ax2 = axes[1]
scatter = ax2.scatter(df['Internet_pct_2024'], df['Ransomware_2024'],
                      s=df['GCI_2024']*8, c=df['GCI_2024'], cmap='RdYlGn', alpha=0.8, edgecolors='black', linewidths=0.5)
plt.colorbar(scatter, ax=ax2, label='GCI Score 2024')
for idx, row in df.iterrows():
    ax2.annotate(idx, (row['Internet_pct_2024'], row['Ransomware_2024']),
                 textcoords='offset points', xytext=(5, 5), fontsize=8)
ax2.set_title('Ransomware Attacks vs Internet Penetration\n(bubble size = GCI score)', fontsize=12, fontweight='bold')
ax2.set_xlabel('Internet Penetration 2024 (%)')
ax2.set_ylabel('Ransomware Incidents 2024')

plt.tight_layout()
plt.savefig('cybersecurity_latam.png', dpi=150, bbox_inches='tight')
plt.show()
````
